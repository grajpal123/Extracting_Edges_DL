## EXTRACTING EDGES

    Here we will be seeing how we can extract the edges form and object using various filters

### `IMPORTING LIBRARIES`

```python
import skimage
from skimage.io import imread, imshow, imsave
from skimage.color import rgb2gray, rgba2rgb
from skimage.filters import prewitt_h, prewitt_v, sobel_h, sobel_v
import numpy as np
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings('ignore')

%matplotlib inline 
```

```python
print('Skimage Version:', skimage.__version__)
```

<br>

### `LOADING IMAGE`

```python
image = imread('dog.jpg', as_gray = True)
print("Shape of image: {}".format(image.shape))
imshow(image)
```

<br>

### `APPLYING KERNELS / FILTERS - SOBEL & PREWITT`

```python
# Capturing features along horizontal axis

edges_sobel_horizontal   = sobel_h(image)
edges_prewitt_horizontal = prewitt_h(image)
```

```python
plt.figure(figsize = (20, 6))

plt.subplot(1,2,1)
imshow(edges_sobel_horizontal, cmap = 'gray')

plt.subplot(1,2,2)
imshow(edges_prewitt_horizontal, cmap = 'gray')
```

    As we can see in sobel_h filter, it is highlighting the horizontal axis eg. head, which is not the case in sobel_v.

```python
# Capturing features along vertical axis 

edges_sobel_vertical   = sobel_v(image)
edges_prewitt_vertical = prewitt_v(image)
```

```python
plt.figure(figsize = (20, 6))

plt.subplot(1,2,1)
imshow(edges_sobel_vertical, cmap = 'gray')

plt.subplot(1,2,2)
imshow(edges_prewitt_vertical, cmap = 'gray')
```

    In sobel_v filter, it is highlighting the vertical features around the neck, nose which is not the case in sobel_h.

<br>

> Now, the images which we got after applying the filter we can just see the object edges and see taht the background noise has been reduced hence reducing the computational power.


> Also, here we have used the kernels or the filters to calculate the edge features by ourself but generally the model learns these features on their own.
